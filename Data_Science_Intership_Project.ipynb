# Task 3: Exploratory Data Analysis (EDA)

In this section , we analyze the datasetto discover patterns, trends, and business insight.

```python
print(df["Product"].value_counts())
```

```python
import matplotlib.pyplot as plt

df["Product"].value_counts().head(10).plot(kind="bar")

plt.title("Top 10 Selling Products")
plt.xlabel("Product")
plt.ylabel("Number of Sales")

plt.show()
```

# Product Analysis Insight

The analysis shows that some products are significantly more popular than others. The highest-selling products contribute the most to total customer purchases.

```python
print(df["PaymentMethod"].value_counts())
```

```python
df["PaymentMethod"].value_counts().plot(kind="pie", autopct='%1.1f%%')

plt.title("Payment Method Distribution")

plt.ylabel("")

plt.show()
```

```python
print("Average Order Price:", df["TotalPrice"].mean())
```

```python
print("Highest Order Price:", df["TotalPrice"].max())
```

```python
print("Lowest Order Price:", df["TotalPrice"].min())
```

```python
print(df["OrderStatus"].value_counts().plot(kind="bar"))

plt.title("Order Status Distribution")

plt.xlabel("Order Status")
plt.ylabel("Count")

plt.show()
```

# Exportory Data Analysis Insights

Key finding from the dataset:

- Some product generated significantly higher sales than others.
- Digital payment methods were commonly used by custumers.
- Customer purchasing trends revealed valuable busness insights.
- Sales distribution varied across different product and order types.

# Task 4: Data Visualization

This section presents visual representations of the dataset to highlight trends, partterns, and business insights.

```python
df["TotalPrice"].plot(kind="hist", bins=10)

plt.title("Sales Distribution ")
plt.xlabel("Total Price")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.scatter(df["Quantity"], df["TotalPrice"])

plt.title("Quantity vs Total Price")
plt.xlabel("Quantity")
plt.ylabel("Total Price")

plt.show()
```

```python
print(df["ReferralSource"].value_counts())
```

```python
df["ReferralSource"].value_counts().plot(kind="bar")

plt.title("Referral Source Distribution")
plt.xlabel("Referral Source")
plt.ylabel("Count")

plt.show()
```

# Visualization Insights

The visualization revealed important trenda within the dataset :

- Certain products generated significatly more than others.
- Customer purchases were concentrated within specific price ranges.
- Digital marketing and referral source conmtributed to customer traffic.
- Higher product quantities generally resulted in higher total price.
- Visual analysis help identify patterns and trends more clearly.
